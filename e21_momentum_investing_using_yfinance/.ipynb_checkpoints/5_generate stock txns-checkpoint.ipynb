```python
import pandas as pd
import dateutil
from pandas.tseries.offsets import Day, BusinessMonthBegin, MonthEnd
```

```python
df_tips = pd.read_csv('out/tips_based_on_past_month.csv')
```

```python
df_tips
```

```python
txn_dates = []
close_l = []

def get_txn_date_and_close(row):    
    txn_date = row['past_month']
    txn_date += '-01'
    txn_date = dateutil.parser.parse(txn_date) + MonthEnd() + Day()
    
    df = pd.read_csv('data/individual stocks/' + row['stock'])
    df['Date '] = df['Date'].apply(lambda x: dateutil.parser.parse(x).strftime("%Y-%m-%d"))
    
    txn_dates.append(txn_date)
    
    j = txn_date + Day()
    close = df[df['Date'] == str(txn_date).split()[0]]['Close'].values
    while len(close) == 0:
        j += Day()
        close = df[df['Date'] == str(j).split()[0]]['Close'].values
        
    close_l.append(close[0])
    
_ = df_tips.apply(get_txn_date_and_close, axis = 1)
```

```python
df_tips['txn_date'] = txn_dates
df_tips['close'] = close_l
df_tips['amt'] = 200
```

```python
df_tips.sort_values(by='txn_date', inplace = True)
```

```python
df_tips.to_csv('out/stock_txns.csv', index = False)
```
