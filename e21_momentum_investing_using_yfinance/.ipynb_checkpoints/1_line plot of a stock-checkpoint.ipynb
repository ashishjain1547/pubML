```python
import pandas as pd
import dateutil
```

```python
df = pd.read_csv('data/individual stocks/TATASTEEL.NS.csv')
```

```python
df['Date'] = df['Date'].apply(lambda x: dateutil.parser.parse(x))
```

```python
df.sort_values(by="Date", inplace = True)
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.plot(df['Date'], df['Close'])
plt.show()
```
