```python
import pandas as pd
import dateutil
from datetime import datetime
import statistics as st
import os
```

```python
df_list = []
stock_names = []
```

```python
for dirpath, subdir, files in os.walk('./data/individual stocks/'): 
    for i in files:
        df = pd.read_csv(dirpath + i, index_col= False)
        df_list.append(df)
        stock_names.append(i)
```

```python
stock_names
```

```python
def get_monthly_slopes(df, stock_name):
    df['month'] = df['Date '].apply(lambda x: str(dateutil.parser.parse(x).year) + "-" 
                                    + str(dateutil.parser.parse(x).month).zfill(2))
    month = df['month'].unique()
    try:
        df['close '] = df['close '].apply(lambda x: float(x.replace(',', '')))
    except AttributeError as e:
        pass
    
    months = []
    slopes = []

    for i in month:
        x = list(df[df['month'] == i].index)
        y = list(df[df['month'] == i]['close '])

        try:
            slope, intercept = st.linear_regression(x, y)
            slopes.append(slope)
            months.append(i)
        except st.StatisticsError as e:
            print(e)

    df_out = pd.DataFrame({
        'month': months,
        stock_name: slopes
    })
    return df_out
```

```python
df_monthly_slopes = [get_monthly_slopes(df = i[0], stock_name = i[1]) for i in zip(df_list, stock_names)]
```

```python
df_rep = df_monthly_slopes[0]
for i in df_monthly_slopes[1:]:
    df_rep = pd.merge(df_rep, i, on = 'month', how = 'inner')
```

```python
df_rep.sort_values(by='month')
```

```python
df_rep.to_csv('out/monthly_slopes.csv', index = False)
```
