```python
import pandas as pd
```

```python
monthly_slopes = pd.read_csv('out/monthly_slopes.csv')
```

```python
monthly_slopes
```

```python
def get_top_five_for_each_month(monthly_slopes):
    months = monthly_slopes['month'].values
    cnt = 0
    
    df_month = []
    df_stock = []
    
    for m in months:
        slopes = monthly_slopes[monthly_slopes['month'] == m].to_dict()
        slopes.pop('month')
        
        stocks_sorted_by_slopes = sorted(slopes.items(), key=lambda x:x[1][cnt], reverse = True)
        cnt += 1
        stocks_sorted_by_slopes = list(dict(stocks_sorted_by_slopes).keys())[:5]
        
        _ = [df_month.append(m) for i in stocks_sorted_by_slopes]
        _ = [df_stock.append(i) for i in stocks_sorted_by_slopes]
        
    df_rep = pd.DataFrame({
        "past_month": df_month,
        "stock": df_stock
    })
    return df_rep
```

```python
df = get_top_five_for_each_month(monthly_slopes)
```
