```python
# Drawing line chart

# 1: Loading the CSV file
import csv
FILE_NAME = 'C:/Users/Ashish Jain/Desktop/BTC.csv'

time_series_list = []
price_list = []

with open(FILE_NAME, newline='') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=',', quotechar='|')
    for row in spamreader:
        time_series_list.append(row[0] + ' ' + row[1])
        price_list.append(row[3])
        
# 2: Generating the line chart
import matplotlib.pyplot as plt
import numpy as np

time_series = np.array(time_series_list)
y_stack = np.array(price_list) 

fig = plt.figure(figsize=(11,8))
ax1 = fig.add_subplot(111)

# Color code 'k' for black
ax1.plot(time_series, y_stack, label='USD/BTC', color='k', marker=',') 

plt.xticks(time_series)
plt.xlabel('Time')

handles, labels = ax1.get_legend_handles_labels()
lgd = ax1.legend(handles, labels, loc='upper center', bbox_to_anchor=(1.15,1))
ax1.grid('on')

plt.savefig('Line Chart.png')
```

```python
# Code for Python 2.6
# v1
import urllib2
from urllib2 import URLError
import json
from time import sleep, strftime
import logging

def getBitcoinPrice():
    URL = 'https://www.bitstamp.net/api/ticker/'
    try:
        response = urllib2.urlopen(URL)
        data = json.load(response)         
        priceFloat = float(data['last']) #Lastprice
        strOutput = data['timestamp'] + "," + str(priceFloat)
        return str(strOutput)
    except URLError, e:
        print("Error querying Bitstamp API: ", e.reason)

LOG_FILENAME = 'BTC.log'
logging.basicConfig(filename=LOG_FILENAME,level=logging.DEBUG)

logging.debug("Date , Time , Unix Timestamp , USD/BTC")
while True:
    line = strftime("%d/%m/%Y") + "," + strftime("%H:%M:%S") + "," + str(getBitcoinPrice())
    logging.debug(line)
    sleep(20)

# Output: 
# DEBUG:root:Date , Time , Unix timestamp , USD/BTC
# DEBUG:root:06/01/2018,02:04:19,1515184421,16384.75
```

```python
# Code for Python 2.6
# v3
# Gets price date and time from Unix timestamp that comes with the JSON response

import urllib2
from urllib2 import URLError
import json
from time import sleep, strftime
import logging
import datetime

def getBitcoinPrice():
    URL = 'https://www.bitstamp.net/api/ticker/'
    try:
        response = urllib2.urlopen(URL)
        data = json.load(response)         
        priceFloat = float(data['last']) #Lastprice
        
        priceDate = datetime.datetime.fromtimestamp(int(data['timestamp'])).strftime('%d/%m/%Y')
        priceTime = datetime.datetime.fromtimestamp(int(data['timestamp'])).strftime('%H:%M:%S')

        return str(priceDate + "," + priceTime + "," + str(priceFloat))
    except URLError, e:
        print("Error querying Bitstamp API: ", e.reason)

LOG_FILENAME = 'BTC.csv'
# URL: https://docs.python.org/3/howto/logging.html
logging.basicConfig(filename=LOG_FILENAME, level=logging.DEBUG, format='%(message)s')

logging.debug("Date,Time,USD/BTC")
while True:
    #print(getBitcoinPrice())
    logging.debug(str(getBitcoinPrice()))
    sleep(20)

# Output: 
# Date,Time,USD/BTC
# 06/01/2018,02:04:19,16384.75
        
```

```python
# Change Microsoft office default formatted dates "dd-mm-yy" to our Python format "dd/mm/yyyy"
import re
f = open('C:/Users/Ashish Jain/Desktop/BTC.csv')
raw = f.read()
linesArr = re.split(r'\r?\n', raw)
out = open('C:/Users/Ashish Jain/Desktop/BTC(2).csv', 'a')
for j in range(0, len(linesArr) - 1):
    if j != 0 and (linesArr[j] == "Date,Time,USD/BTC" or "None" in linesArr[j]):
        continue
    printStr = linesArr[j][0:] + "\n"
    if "-" in linesArr[j]:
        printStr = linesArr[j][0:2] + '/' +  linesArr[j][3:5] + '/20' + linesArr[j][6:] + "\n"
    out.write(printStr)
#    04-03-18,22:18:40,11107.55
out.close()
```
